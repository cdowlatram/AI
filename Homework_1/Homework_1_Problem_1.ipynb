```
from mpl_toolkits import mplot3d
```

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Local Minimum

```
def f(x, y):
    return (x ** 2 + y ** 2)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot3D([0], [0], [0], c='r', marker='o', label='Local Minimum (0, 0, 0)')
ax.legend();
```
