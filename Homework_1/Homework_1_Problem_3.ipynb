```
%matplotlib inline
import matplotlib.pyplot as plt
```

## Loading the MNIST digits data set

```
from keras.datasets import mnist

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

## Computing Average Digit

```
for i in range(0, 10):
  train = train_images[train_labels == i].astype(float)
  test = test_images[test_labels == i].astype(float)
  avg = (sum(train) + sum(test))/(len(train) + len(test))
  plt.title(i)
  plt.imshow(avg)
  plt.show()
  plt.close()
```
